```python
def remove_duplicates_df(self):
    """Removing duplicates from the category file just to be safe."""
    input_file_path = r"business collector\{}_{}.csv".format(self.catergory, self.town_center)
    output_file_path = r"business collector\{}_{}_cleaned.csv".format(self.catergory, self.town_center)

    df_rd = pd.read_csv(input_file_path)
    df_rd.set_index("name", inplace=True)
    df_rd = df_rd.drop_duplicates()

    df_rd.to_csv(output_file_path, mode="w", header=True)
    print("Removing duplicates from CSV. Cleaned data saved to:", output_file_path)
```
